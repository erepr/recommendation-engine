```python
def set_min_budget_fixedprice(x):
    if(len(x) > 0):
        val = x[0].lstrip(' ')
        finalval = 1
        if(val == 'Under $250'):
            finalval = 125
        elif(val == '$250-$500'):
            finalval = 250
        elif(val == '$500-$1k'):
            finalval = 500
        elif(val == '$1k-$2.5k'):
            finalval = 1000
        elif(val == '$2.5k-$5k'):
            finalval = 2500
        elif(val == '$5k-$10k'):
            finalval = 5000
        elif(val == '$10k-$25k'):
            finalval = 10000
        elif(val == 'Over $25k'):
            finalval = 25000
        else:
            finalval = val
    else:
        finalval = np.nan
    return finalval
```

```python
df_fixed_price = df[df.Payment_type == 'Fixed Price'].Budget.map(set_min_budget_fixedprice)
```

```python
df.head()
```

```python
def get_min_hours(x):
    if(x == '1-10 hrs/wk'):
        return 5/5
    elif(x == '10-30 hrs/wk'):
        return 10/5
    elif(x == '30+ hrs/wk'):
        return 30/5

def get_min_duration(x):
    if(x == '1-5 days'):
        return 2
    elif(x == '1-4 weeks'):
        return 5
    elif(x == '1-3 months'):
        return 30
    elif(x == '3-6 months'):
        return 90
    elif(x == '6+ months'):
        return 180
    
    
def set_min_budget_hourly(x):
    if(len(x) > 2):
        min_max = x[0].strip()
        if(min_max == 'No rate specified'):
            return np.nan
        else:
            if 'Min' in min_max:
                minRate = min_max.replace('Min $', '').replace(',', '')
            elif 'Max' in min_max:
                minRate = min_max.replace('Max $', '').replace(',', '')
            else:
                minRate = list(map(lambda x: x.lstrip('$').replace(',', '') , min_max.split(' - ')))[0]
            minHours = get_min_hours(x[1].strip())
            minDuration = get_min_duration(x[2].strip())
            min_budget = int(minRate) * minHours * minDuration
            return min_budget
    else:
        return np.nan
df_hourly = df[df.Payment_type == 'Hourly'].Budget.map(set_min_budget_hourly)
```

# Run Recommendation Algo and evaluate final result

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import sigmoid_kernel
```

```python
df['Skills_str'] = df.Skills.map(lambda x: ', '.join(x))
```

```python
def get_similar_jobs(jobId, w_title, w_desc, w_skills, num_similar_jobs):
    #jobId = 1737717
    df_clustered = df[df['cluster'] == df[df['Job_id'] == jobId].cluster.values[0]]
    df_clustered = df_clustered.reset_index(drop = True)
    jobIndex = df_clustered[df_clustered['Job_id'] == jobId].index[0]
    title =  df_clustered[df_clustered['Job_id'] == jobId].Job_title
    desc = df_clustered[df_clustered['Job_id'] == jobId].Job_description
    skills = df_clustered[df_clustered['Job_id'] == jobId].Skills_str
    w_title = w_title
    w_desc = w_desc
    w_skills  = w_skills  
    #Title
    tfv_title = TfidfVectorizer(min_df =  1, ngram_range= (1,2), max_features = None, stop_words = 'english', 
                     strip_accents = 'unicode', analyzer = 'word', token_pattern=r'\w{1,}')
    tfv_matrix_title = tfv_title.fit_transform(df_clustered['Job_title'])
    sigmoid_title = sigmoid_kernel(tfv_matrix_title, tfv_matrix_title)
    sigmoid_title = sigmoid_title[jobIndex]
    wtd_sigmoid_title=w_title * sigmoid_title
    wtd_sigmoid_title

    #Description
    tfv_desc = TfidfVectorizer(min_df =  3, ngram_range= (1,3), max_features = None, stop_words = 'english', 
                     strip_accents = 'unicode', analyzer = 'word', token_pattern=r'\w{1,}')
    tfv_matrix_desc = tfv_desc.fit_transform(df_clustered['Job_description'])
    sigmoid_desc = sigmoid_kernel(tfv_matrix_desc, tfv_matrix_desc)
    sigmoid_desc = sigmoid_desc[jobIndex]
    wtd_sigmoid_desc=w_desc * sigmoid_desc

    #Skills
    tfv_skills = TfidfVectorizer(min_df =  3, ngram_range= (1,2), max_features = None, stop_words = 'english', 
                     strip_accents = 'unicode', analyzer = 'word', token_pattern=r'\w{1,}')
    tfv_matrix_skills = tfv_skills.fit_transform(df_clustered['Skills_str'])
    sigmoid_skills = sigmoid_kernel(tfv_matrix_skills, tfv_matrix_skills)
    sigmoid_skills = sigmoid_skills[jobIndex]
    wtd_sigmoid_skills=w_desc * sigmoid_skills


    sigmoid_total= wtd_sigmoid_title + wtd_sigmoid_desc + wtd_sigmoid_skills 
    
    job_scores = sorted(list(enumerate(sigmoid_total)), key= lambda x: x[1], reverse = True)
    df_indices = [i[0] for i in job_scores]
    
    return df_clustered['Job_id'].iloc[df_indices].iloc[1:(num_similar_jobs + 1)]
```
