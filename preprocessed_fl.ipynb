```python
import pandas as pd
import numpy as np
import re
import statistics 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data.csv")
df
```

```python
ate = []
for i in range(len(df)):
    if "M" in df['All-Time Earnings'].iloc[i]:
        num = re.findall("\d+\.\d+", df['All-Time Earnings'].iloc[i])
        sum = float(num[0]) * 1000000
        ate.append(sum)
    elif "K" in df['All-Time Earnings'].iloc[i]:
        num = re.findall("\d+", df['All-Time Earnings'].iloc[i])
        sum = int(num[0]) * 1000
        ate.append(sum)
    else:
        ate.append(int(re.sub('[^0-9,]', "", df['All-Time Earnings'].iloc[i]).replace(",", "")))
```

```python
df['New_All_Time_Earning'] = ate
```

```python
new_transactions_complete = []
for i in range(len(df)):
    new_transactions_complete.append(int(re.sub('[^0-9,]', "", df['Transactions Completed'].iloc[i]).replace(",", "")))
    
df['New_Transactions_Complete'] = new_transactions_complete
```

```python
df.drop(['All-Time Earnings', 'Transactions Completed'], axis = 1)
df
```

```python
df['Average Earnings'] = round(df['New_All_Time_Earning'] / df['New_Transactions_Complete'], 2)
df['Average Earnings']
```

```python
df['Starting Rate']=df['Starting Rate'].replace(np.nan, "0")
df['Hourly Rate']=df['Hourly Rate'].replace(np.nan, "0")
```

```python
temp1 = {'Min_Starting_Rate': [],
         'Max_Starting_Rate': [],
         'Min_Hourly_Rate': [],
         'Max_Hourly_Rate': []}

for i in range(len(df)):
    ax = re.findall("\d+", df['Starting Rate'].iloc[i])
    bx = re.findall("\d+", df['Hourly Rate'].iloc[i])
    test_list1 = [int(j) for j in ax]
    test_list2 = [int(j) for j in bx]
    temp1['Min_Starting_Rate'].append(min(test_list1))
    temp1['Max_Starting_Rate'].append(max(test_list1))
    temp1['Min_Hourly_Rate'].append(min(test_list2))
    temp1['Max_Hourly_Rate'].append(max(test_list2))
```

```python
df2 = pd.DataFrame(temp1)
df2
```

```python
result = pd.concat([df, df2], axis=1)
result
```

```python
result = result.drop(['Hourly Rate', 'Starting Rate', 'Freelancer_Link'], axis = 1)
result
```

```python
cx = result[['Total_Feedback_Received', 'Transactions Completed', 'Employers']]
cx=cx.replace(np.nan, 0)
cx
```

```python
result['Emp_to_Feedback_Ratio'] = cx['Employers'] / cx['Total_Feedback_Received']
result=result.replace(np.inf, 0)
```

```python
a1 = result['Rating'].replace(np.nan, 0)

weighted_rating = []

for i in range(len(df)):
    if type(a1.iloc[i]) == str:
        weighted_rating.append(float(a1.iloc[i].strip('%')) / 20)
    else:
        weighted_rating.append(a1.iloc[i])
        
result['Weighted_Rating'] = weighted_rating
```
